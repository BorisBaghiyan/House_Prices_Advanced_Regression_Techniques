```python
df_test = pd.read_csv('test.csv')
```

```python
df_train = df_train.drop('Id',axis = 1)
df_test = df_test.drop('Id',axis = 1)
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
label = LabelEncoder()
```

```python
for i in df_train:
    if i == 'LotFrontage':
        df_train['LotFrontage'] = df_train['LotFrontage'].fillna(df_train['LotFrontage'].median())
    elif df_train[i].dtype == object:
        df_train[i] = label.fit_transform(df_train[i])
```

```python
df_test.info()
```

```python
df_test['LotFrontage'] = df_test['LotFrontage'].fillna(df_test['LotFrontage'].median())
```

```python
df_test['MSZoning'] = df_test['MSZoning'].fillna(np.random.choice(a = ['RL','RM','FV','C (all)','RH'], p = [0.5,0.2,0.1,0.1,0.1]))
```

```python
df_test['SaleType'] = df_test['SaleType'].fillna('WD')
```

```python
for i in df_test:
    if df_test[i].dtype == "O":
        df_test[i] = label.fit_transform(df_test[i])
```

```python
df_train = df_train.drop('MasVnrArea',axis = 1)
df_test = df_test.drop('MasVnrArea',axis = 1)
```

```python
df_train['GarageYrBlt'] = df_train['GarageYrBlt'].fillna(df_train['GarageYrBlt'].mean())
df_test['GarageYrBlt'] = df_test['GarageYrBlt'].fillna(df_test['GarageYrBlt'].mean())
```

```python
df_test
```

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
model = RandomForestRegressor(
    n_estimators=200,         
    max_depth=10,             
    min_samples_split=5,      
    min_samples_leaf=2,       
    max_features='sqrt',      
    random_state=42,          
    n_jobs=-1                 
)
```

```python
X_train = df_train.drop('SalePrice',axis = 1)
y_train = df_train['SalePrice']
X_test = df_test.copy()
```

```python
model.fit(X_train,y_train)
```

```python
X_test =df_test.fillna(0)
```

```python
y_pred = model.predict(X_test)
```

```python
sub = pd.read_csv('sample_submission.csv')
```

```python
sub['SalePrice'] = y_pred
```

```python
sub.to_csv('sub.csv',index=False,index_label=False)
```
